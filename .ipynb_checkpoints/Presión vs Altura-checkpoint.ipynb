# <center>DEPENDENCE OF ATMOSPHERIC PRESSURE WITH ALTITUDE</center>


###  <center>P = 1013.25 * (1 - 0.0000225577 * H)^5.2559</center>
                                                            -'Internacional Standard Atmosphere.'
                                                            - By Nikolay Murillo.

```python
# Import libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
# Code

H = list(range(0,22000,2850)) #altura
presion = []
for i in H:
    presion.append(1013.25 * ((1 - 0.0000225577 * i) ** 5.2559) / 10)
df = pd.DataFrame({'Altura [m]':H,'Presión [kPa]':presion})
df = df.round(2)
df
```

```python
fig = plt.figure(figsize =(10,7))
plt.plot(H,presion, color = 'g', linestyle='-', marker='o',)
plt.xlabel('Altitude [m]')
plt.xlim(0,20000)
plt.ylim(0,102)
plt.ylabel('Pressure [kPa]')
plt.title('DEPENDENCE OF ATMOSPHERIC PRESSURE WITH ALTITUDE')
fig.tight_layout()
plt.show()
```
